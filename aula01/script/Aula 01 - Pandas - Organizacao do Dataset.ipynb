## Informações Gerais do DataFrame

O método `.head()` é ótimo para uma primeira impressão. Mas para um resumo mais técnico, usamos o `.info()`. Para saber linhas e colunas, o atributo `shape` (cuidado é atributo, não método)

Este método nos mostra:
- O número total de linhas (entradas).
- O número total de colunas.
- O nome de cada coluna.
- A quantidade de valores **não-nulos** em cada coluna (muito útil para identificar dados faltantes!).
- O tipo de dado de cada coluna (ex: `int64` para números inteiros, `object` para textos, `float64` para números com casas decimais).
- A quantidade de memória que o DataFrame está usando.


```python
df.shape
```

```python
df.info()
```

### Mudando valores das colunas

A coluna '2.h_faixa_salarial' tem valore em texto. Sem perder (muito) a semântica, é possível alterar para valores numéricos.



```python
df['2.h_faixa_salarial'].value_counts()
```

```python
# Dicionário pra mapear os valores para transformação
dict_mapeamento = {
    'de R$ 8.001/mês a R$ 12.000/mês': 10000.0,
    'de R$ 12.001/mês a R$ 16.000/mês': 14000.0,
    'de R$ 6.001/mês a R$ 8.000/mês': 9000.0,
    'de R$ 4.001/mês a R$ 6.000/mês': 5000.0,
    'de R$ 16.001/mês a R$ 20.000/mês': 18000.0,
    'de R$ 3.001/mês a R$ 4.000/mês': 3500.0,
    'de R$ 20.001/mês a R$ 25.000/mês': 22500.0,
    'de R$ 2.001/mês a R$ 3.000/mês': 2500.0,
    'de R$ 25.001/mês a R$ 30.000/mês': 27500.0,
    'de R$ 30.001/mês a R$ 40.000/mês': 35000.0,
    'de R$ 1.001/mês a R$ 2.000/mês': 1500.0,
    'Acima de R$ 40.001/mês': 40000.0,
    'Menos de R$ 1.000/mês': 500.0,
}


df['faixa_salarial_numerico'] = df['2.h_faixa_salarial'].map(dict_mapeamento)

df['faixa_salarial_numerico'].value_counts()
```

### Renomear colunas

As colunas tem nomes muito grandes e na hora de fazer a análise pode atrapalhar. Vamos renomear algumas colunas! Vamos criar nomes mais curtos e significativos para as colunas que vamos usar. Não precisamos renomear todas agora, podemos focar em algumas principais.

Para renomear, usamos o método `.rename()`, passando um **dicionário** onde a chave é o nome antigo e o valor é o nome novo.


```python
df.head(1)
```

```python
# Vamos pegar as primeiras colunas como exemplo.
# O nome original é muito grande, então vamos copiá-lo da saída da célula anterior.

colunas_para_renomear = {
    '0.a_token': 'token',
    '0.d_data/hora_envio': 'timestamp_envio',
    '1.a.1_faixa_idade': 'faixa_etaria',
    '1.b_genero': 'genero',
    '1.c_cor/raca/etnia': 'etnia',
    '1.a_idade': 'idade',
    '1.d_pcd': 'pcd',
    '1.e_experiencia_profissional_prejudicada': 'xp_profissional_prejudicada',
    '1.i.1_uf_onde_mora': 'uf_residencia',
    '1.l_nivel_de_ensino': 'nivel_ensino',
    '1.m_área_de_formação': 'area_formacao',
    '2.a_situação_de_trabalho': 'situacao_trabalho',
    '2.f_cargo_atual': 'cargo_atual',
    '2.h_faixa_salarial': 'faixa_salarial',
    '2.i_tempo_de_experiencia_em_dados': 'tempo_experiencia_dados'
}

df.rename(columns=colunas_para_renomear, inplace=True)

df.head()
```
